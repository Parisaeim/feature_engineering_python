```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import FunctionTransformer
```

```python
data = pd.read_csv('BostonHousing.csv')
```

```python
data.head()
```

```python
# تبدیل ریشه مکعب / مربع با Numpy:
data_tf = data.copy()
```

```python
# اعمال تبدیل ریشه مربع به مجموعه‌ای از متغیرها 
data_tf[['lstat','nox','dis','rm']]=np.sqrt(data[['lstat','nox','dis','rm']])
```

```python
# اعمال تبدیل ریشه سوم به مجموعه‌ای از متغیرها
data_tf[['lstat','nox','dis','rm']]=np.cbrt(data[['lstat','nox','dis','rm']])
```

```python
# تبدیل ریشه مربع / مکعب با scikit-learn:
transformer = FunctionTransformer(np.sqrt, validate=True)
```

```python
cols = ['lstat','nox','dis','rm']
data[cols] = data[cols].fillna(0.1)
data_tf = transformer.transform(data[cols])
```

```python
data_tf = pd.DataFrame(data_tf, columns=cols)
```
