```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import FunctionTransformer
from feature_engine.transformation import ReciprocalTransformer
```

```python
data = pd.read_csv('BostonHousing.csv')
```

```python
data.head()
```

```python
# تبدیل معکوس با Numpy:
# کپی گرفتن از داده تا بتوانیم متغیرهای تغییر یافته را در آن ذخیره کنیم
data_co = data.copy()
```

```python
data_co[['lstat', 'nox', 'dis', 'rm']] = np.reciprocal(data[['lstat', 'nox', 'dis', 'rm']])
```

```python
#تبدیل معکوس با Scikit_learn:
transformer = FunctionTransformer(np.reciprocal, validate=True)
```

```python
#ساختن لیستی از متغیرها که تبدیل روی آنها انجام شود
cols= ['lstat', 'nox', 'dis', 'rm']
data[cols] = data[cols].fillna(0.1) 
data_tf = transformer.transform(data[cols])
```

```python
#تبدیل آرایه نامپای برگشتی به دیتافریم پانداس
data_tf = pd.DataFrame(data_tf, columns=cols)
```

```python
#تبدیل معکوس با Feature_engine
rt = ReciprocalTransformer(variables = ['lstat', 'nox', 'dis', 'rm'])
rt.fit(data)
```

```python
data_tf = rt.transform(data)
```
